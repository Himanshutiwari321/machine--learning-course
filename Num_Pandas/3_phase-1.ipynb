##numpy array and basic 

```python
import numpy as np 
```

##creating array from list 

```python
arr_1d = np.array( [1,2,3,4,5,6] )
print ("1D array:", arr_1d)

arr_2d = np.array( [[1,2,3], [4,5,6]] )
print ("2D array:", arr_2d)
```

##list vs numpy array 

```python
py_list =[1,2,3]
print ("python list multiplication",py_list *2)

np_array = np.array([1,2,3])
print ("python array multiplication",np_array *2)

import time 
start = time.time()
py_list = [i*2 for i in range(1000000)]
print ("\n list operation time :",time.time()-start)

start = time.time ()
np_array = np.arange(1000000)*2
print ("\n numpy operatin time :", time.time()-start)
```

##Creating array from scartch

```python
zeros = np.zeros((3,4))
print("zeros array \n ", zeros )

one = np.ones((3,4))
print ("ones array:\n", one)

full = np.full((2,2),7)
print ("full array:\n",full)

random = np.random.random((2,3))
print ("random array:",random)

sequence = np.arange(0,10,2)
print("sequence array \n :" , sequence )


```

##VECTOR , METRIX , TENSOR 

```python
vector = np.array ([1,2,3])
print ("Vector :", vector )

metrix = np.array ([[1,2,3],[4,5,6]])
print (" \n Metrix: ", metrix )

tensor = np.array ([[[1,2],[3,4]],[[4,5],[6,7]]])
print ("\n Tensor :", tensor )
```


##Array properties 

```python
arr = np.array ([[1,2,3],[4,5,6]])
print ("shape ", arr.shape)
print ("Dimension", arr.ndim)
print ("Size", arr.size)
print ("Dtype ",arr.dtype)
```

Array reshaping 

```python
arr = np.arange (12)
print ("original array", arr)

reshaped = arr.reshape ((3,4))
print (" \n Peshaped array",reshaped)

flattend = reshaped.flatten()
print ("\n flattend array ", flattend )

#raval (returns view , instead of copy )
raveled = reshaped.ravel()
print ("\n raveled array ", raveled)

#transpose 
transpose = reshaped.T
print ("\n Transposed array ", transpose)


```
