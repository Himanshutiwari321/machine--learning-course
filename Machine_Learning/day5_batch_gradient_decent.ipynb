### batch gradient decent 

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split


X, y =load_diabetes(return_X_y=True)
```

```python
# Feature names bhi nikalna
data = load_diabetes()
feature_names = data.feature_names

# Numpy array ko DataFrame me convert karna
df = pd.DataFrame(X, columns=feature_names)

# Target column add karna
df["target"] = y

# First 5 rows dekhna
df.head()
```

```python
# X = inputs (all columns except target)
X = df.drop("target", axis=1)


# y = output (only target column)
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2)
```

```python
class GDRegressor:
     
    def __init__(self, learning_rate=0.01, epochs=100):
        self.m = None 
        self.b = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X_train, y_train):
        self.b = 0 
        self.m = np.ones(X_train.shape[1])   
        

        for i in range(self.epochs):
            y_hat = np.dot(X_train, self.m) + self.b
            b_der = -2 * np.mean(y_train - y_hat) 
            self.b = self.b-(self.lr*b_der)

            m_der = -2 * np.dot((y_train - y_hat), X_train) / X_train.shape[0]  
            self.m =self.m -(self.lr*m_der)
            
           

    def predict(self, X_test):
        return np.dot(X_test, self.m) + self.b

```

```python
gd= GDRegressor(epochs=1000,learning_rate=0.5)

gd.fit(X_train, y_train)
  
    


  
```

```python
y_pred =gd.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
