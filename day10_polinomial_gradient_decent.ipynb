```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression , SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler

from sklearn.metrics import r2_score 
```

```python
# 3D polynomial regression
x = 7 * np.random.rand(100, 1) - 2.8
y = 7 * np.random.rand(100, 1) - 2.8

z = x**2 + y**2 + 0.2*x + 0.2*y + 0.1*x*y +2 + np.random.randn(100, 1)
# z = x^2 + y^2 + 0.2x + 0.2y + 0.1xy + 2
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
class MeraPoly:
    
    def __init__(self , degree = 2):
        self.m = None
        self.b = None 
        self.degree = degree 
 
#transform X_train to X_poly 

    def fit(self , X_train , y_train):

        X_poly = np.ones((X_train.shape[0] , 1))
        for i in range (1 , self.degree+1):
            X_poly = np.hstack ((X_poly , X_train**i))

        betas = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(y_train)
        self.intercept_ = betas[0]
        self.coef_ = betas[1:]


    
    def predict(self , X_test):

        X_test_poly = np.ones((X_test.shape[0] , 1))
        for i in range (1 , self.degree+1):
            X_test_poly = np.hstack((X_test_poly , X_test**i))


        y_pred = np.dot (X_test_poly , self.m) + self.b
        return y_pred 
 



              

        
```

```python
model = MeraPoly(degree=2)
model.fit(X_train, y_train)




```
