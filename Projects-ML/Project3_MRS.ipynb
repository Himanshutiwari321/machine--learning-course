```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movie = movies.merge(credits, on='title')
```

```python
movie.head(1)
movie.info()
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
movie_s=movie[['movie_id','title','overview','genres','keywords','cast','crew']]
movie_s.head()
```
