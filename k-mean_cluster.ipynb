### k- means clustering 2d data

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("student_clustering.csv")
df.shape 
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt 
plt.scatter(df["cgpa"],df["iq"])
```

### step-1

```python
from sklearn.cluster import KMeans
```

```python
wcss =[]

for i in range (1,11):
 km =KMeans(n_clusters=i)
 km.fit_predict(df)
 wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

### step-2

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)

y_means

```

```python
X[y_means==3,0],X[y_means==3,1]
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],color="blue")
plt.scatter(X[y_means==1,0],X[y_means==1,1],color="red")
plt.scatter(X[y_means==2,0],X[y_means==2,1],color="green")
plt.scatter(X[y_means==3,0],X[y_means==3,1],color="yellow")
```

```python
### red = cgpa and iq both are high ---- smart and hardworking 
### blue= cgpa high but iq is low ----- means hardworking student 
### yellow =iq high but cgpa is low ----- means smart but not utilize their smartness
### green low iq and low cgpa ------- will have doing work hard
```

### k -means on 3-d data 


```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
X
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0],y=X[:,1], z=X[:,2]) #example --- x=X[:,0] sari row s lekin colum sirf 0 idex wala 
fig.show()
```

```python
wcss =[]

for i in range (1,11):
 km =KMeans(n_clusters=i)
 km.fit_predict(X)
 wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()   

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

```python
df.head()
```

```python
grouped = df.groupby('label')
```

```python
cluster_2 = grouped.get_group(2)
print(cluster_2) # cluster data 

#print(cluster_2.iloc[19]) #  specific row data in cluster
```

```python
fig = px.scatter_3d(df,x='col1', y='col2', z='col3',color='label')
fig.show()
```
